# 词云生成器命令行参数测试

本notebook用于测试main.py中不同命令行参数组合的运行效果。

```python
import sys
import argparse
from pathlib import Path
from main import main
from config import DOCS_DIR
```

## 测试参数解析函数

```python
def test_args(args_list):
    """测试不同命令行参数组合"""
    # 备份原始参数
    orig_argv = sys.argv
    
    try:
        # 设置测试参数
        sys.argv = ['main.py'] + args_list
        print(f"\n测试参数: {' '.join(sys.argv)}")
        
        # 运行主函数
        main()
        
    except Exception as e:
        print(f"运行出错: {e}")
    finally:
        # 恢复原始参数
        sys.argv = orig_argv
```

## 测试默认参数

```python
# 使用默认参数运行
test_args([])
```

## 测试LLM方法(Deepseek)

```python
# 使用Deepseek API
test_args([
    '--method', 'llm',
    '--api-type', 'deepseek',
    '--docs-dir', str(DOCS_DIR),
    '--mask', 'masks/cloud.png'
])
```

## 测试LLM方法(Ollama)

```python
# 使用Ollama API
test_args([
    '--method', 'llm', 
    '--api-type', 'ollama',
    '--docs-dir', str(DOCS_DIR),
    '--mask', 'masks/circle.png'
])
```

## 测试TF-IDF方法

```python
# 使用TF-IDF方法
test_args([
    '--method', 'tfidf',
    '--docs-dir', str(DOCS_DIR),
    '--mask', 'masks/square.png'
])
```

## 测试自定义文档目录

```python
# 指定其他文档目录
custom_docs_dir = Path('custom_docs')
test_args([
    '--method', 'tfidf',
    '--docs-dir', str(custom_docs_dir)
])
```
